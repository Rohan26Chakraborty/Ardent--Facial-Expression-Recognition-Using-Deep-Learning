```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# fetching training data from image data to numpy array format

x_train = []
y_train = []
train_path = 'dataset/fer2013/train/'
expressions = os.listdir(train_path)
for i in range(len(expressions)):
    img_path = [train_path + expressions[i] + '/' + path for path in os.listdir(train_path + expressions[i])]
    img_set = [plt.imread(image) for image in img_path]
    x_train = x_train + img_set
    y_train = y_train + [i] * len(img_path)
    print(f'{expressions[i]} : {len(img_path)} images')

x_train = np.array(x_train)
y_train = np.array(y_train)
print(f'x_train.shape = {x_train.shape}     y_train.shape = {y_train.shape}')
```

```python
# normalizing training data

x_train_normalized = x_train/255.0
x_train_normalized.shape
```

```python
# randomly shuffle training data

indices = np.arange(x_train_normalized.shape[0])
np.random.shuffle(indices)
x_train_shuffle = x_train_normalized[indices]
y_train_shuffle = y_train[indices]

x_train_normalized = x_train_shuffle
y_train = y_train_shuffle
```

```python
# visualising some images

plt.figure(figsize=(12,7))
for i in range(50):
    plt.subplot(5,10,i+1)
    plt.axis('off')
    plt.title(expressions[y_train[i]])
    plt.imshow(x_train_normalized[i], cmap='gray')
```

```python
# fetching testing data from image data to numpy array format

x_test = []
y_test = []
test_path = 'dataset/fer2013/test/'
expressions = os.listdir(test_path)
for i in range(len(expressions)):
    img_path = [test_path + expressions[i] + '/' + path for path in os.listdir(test_path + expressions[i])]
    img_set = [plt.imread(image) for image in img_path]
    x_test = x_test + img_set
    y_test = y_test + [i] * len(img_path)
    print(f'{expressions[i]} : {len(img_path)} images')

x_test = np.array(x_test)
y_test = np.array(y_test)
print(f'x_test.shape = {x_test.shape}     y_test.shape = {y_test.shape}')
```

```python
# normalizing testing data

x_test_normalized = x_test/255.0
x_test_normalized.shape
```

```python
# save the fetched data for future use

path = 'data/'
np.save(path+'fer-trainx-complete', x_train_normalized)
np.save(path+'fer-trainy-complete', y_train)
np.save(path+'fer-testx-complete', x_test_normalized)
np.save(path+'fer-testy-complete', y_test)
```
